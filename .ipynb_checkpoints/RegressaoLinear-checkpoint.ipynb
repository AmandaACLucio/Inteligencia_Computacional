## Importando as bibliotecas


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

%matplotlib inline
```

## Carregando os dados


```python
usedCars = pd.read_csv('cclass.csv')
```

## As variáveis preditoras e a variável de resposta

Para treinar o modelo de regressão, primeiro precisaremos dividir nossos dados em uma matriz **X** que contenha os dados das variáveis preditoras e uma matriz **y** com os dados da variável de destino, nesse caso a coluna Preço. Vamos descartar a coluna Endereço, pois ela possui apenas informações de texto que o modelo de regressão linear não pode usar.

### Matrizes X e y

```python
X = usedCars[['year', 'mileage','engineSize']]
y = usedCars['price']
```

## Relação entre as variáveis preditoras


**Gráficos simples de dispersão**

Pelos gráficos abaixo percebemos que as variáveis preditoras 'year' e 'mileage' possui uma relação linear com menor dispersão dos dados em relação a nossa variável de resposta 'Price'.

```python
sns.pairplot(usedCars)
plt.show()
```

## Separando os dados de treinamento e de validação

Agora vamos dividir os dados em um conjunto de treinamento e um conjunto de testes. Vamos treinar o modelo no conjunto de treinamento, em seguida, usar o conjunto de teste para validar o modelo.

Em nosso exemplo iremos separar de forma randômica 33% dos dados para validação. Estes dados não serão utilizados para determinação dos coeficientes preditores do modelo. 


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

## Criando o Modelo de Regressão Linear Multivariável

Temos como expressão para y:

**y = β0 + β1 × x1 + β2 × x2 + ... + β5 × x5**

Sendo ***xn*** as variáveis preditoras e ***βn*** o coeficiente destas variáveis.


Desta forma, temos o seguinte modelo para estimar o preço do veículo:


**Price = β0 + β1 × 'year' + β2 ×'mileage' + β3 × 'engineSize'**


### Regressão Linear dos mínimos quadrados

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

# Avaliando o modelo

Vamos avaliar o modelo verificando seus coeficientes e como podemos interpretá-los.

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

Fixando o valor das demais variáveis:

- Aumentando 1 unidade de **'Year'** tem-se um incremento no preço do carro de \$1847.87.
- Aumentando 1 unidade de **'Mileage'** tem-se uma dimuição no preço do carro de \$0.14.
- Aumentando 1 unidade de **'Engine Size'** tem-se um incremento no preço do carro de \$8717.33.



# Avaliando o condicionamento

## Outliers

Outliers é uma observação que apresenta um grande afastamento das demais da série, ou que é inconsistente. Podemos observar alguns outliers abaixo:

<img src='imagens/outliers.png' /></a>

```python
from scipy import stats

df = usedCars[['year', 'price', 'mileage', 'engineSize']]

(m, b) = np.polyfit(df['year'], df['price'], 1)

yp = np.polyval([m, b], df['year'])
plt.plot(df['year'], yp, color='r')
plt.grid(True)
plt.scatter(df['year'],df['price'])
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Dados com outliers')
ax = plt.gca()
ax.set_xlim([1989, 2022])
ax.set_ylim([-2000, 90000])

plt.show()

cols = ['year', 'price', 'mileage', 'engineSize']

Q1 = df[cols].quantile(0.035)
Q3 = df[cols].quantile(0.965)
IQR = Q3 - Q1

df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

yp = np.polyval([m, b], df['year'])
plt.plot(df['year'], yp, color='r')
plt.grid(True)
plt.scatter(df['year'],df['price'])
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Dados sem outliers')
ax = plt.gca()
ax.set_xlim([1989, 2022])
ax.set_ylim([-2000, 90000])

plt.show()
```

## Como validar o modelo?

### Validação cruzada - K-Fold

Os dados de treinamento serão divididos em conjuntos de treinamento e teste, um para treinar o modelo e outro para testar o modelo. Após isso, avaliamos o desempenho do modelo com base em uma métrica de erro, determinando assim a precisão do modelo. Esse método, no entanto, não é muito confiável, pois a precisão obtida para um conjunto de testes pode ser muito diferente da precisão obtida para um conjunto de testes diferente. Uma alternativa é a validação cruzada, através da divisão dos dados em seções e alternância entre o modelo de teste do momento.

<img src='imagens/k_fold.png' />

## Cross Validation ou Validação Cruzada

Faremos uma validação cruzada para verificar esse comportamento. Os dados serão divididos em 5 grupos.

```python
from sklearn.model_selection import KFold
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
scores = []
best_lm = LinearRegression()
kf = KFold(n_splits=10, random_state=30, shuffle=True)
```

```python
for train_index, test_index in kf.split(X):
    
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    
    best_lm.fit(X_train,y_train)
    predictions = best_lm.predict(X_test)
     
    MAE = metrics.mean_absolute_error(y_test, predictions)
    print("MAE: ",MAE)
    print('MAE_NORMALIZADO:', MAE/y.mean())
    
    resultado = cross_val_score(best_lm, X_test, y_test, cv = 5)
    print("SCORE: ",resultado.mean())
    
    coeff_df = pd.DataFrame(best_lm.coef_,X.columns,columns=['Coefficient'])
    print(coeff_df,"\n\n\n\n\n ")
    
```
